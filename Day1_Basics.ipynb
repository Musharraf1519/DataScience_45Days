```python
# What is Pandas?
# Pandas is a Python library (a set of pre-written code) that helps you work easily with tables of data ‚Äî like Excel sheets or CSV files. 
# It lets you load, explore, filter, and change data with simple commands.

# Step 1: Import Pandas
# Before you can use Pandas, you need to import it ‚Äî this means loading the library so Python knows you want to use it.

import pandas as pd

# Here, pd is just a short name to call Pandas quickly.

# Step 2: Create a DataFrame (Table)
# A DataFrame is like a spreadsheet or table. You can create one from a dictionary (like a set of key:value pairs).

data = {
    "Name": ["John", "Sara", "Mike"],
    "Age": [23, 29, 31],
    "City": ["NY", "LA", "Chicago"]
}
df = pd.DataFrame(data)
print(df)
# What this does:
# 	‚Ä¢ The data dictionary has 3 keys (Name, Age, City), each with a list of values.
# 	‚Ä¢ pd.DataFrame(data) turns this dictionary into a table.
# 	‚Ä¢ print(df) shows the table.
# Output looks like:
# 	Name	Age	City
# 0	John	23	NY
# 1	Sara	29	LA
# 2	Mike	31	Chicago


# Step 3: Basic DataFrame Operations

# Show the first rows
print(df.head())  
# Shows the first 5 rows by default.

# Filter rows where Age > 25

print(df[df['Age'] > 25])

# Output:
# 	Name	Age	City
# 1	Sara	29	LA
# 2	Mike	31	Chicago


# Add a new column
df['Age_next_year'] = df['Age'] + 1
print(df)
# Adds a new column with age increased by 1.

# Practice for you:
# 	‚Ä¢ Add a column called 'Age_category' that is "Young" if Age < 30 else "Old"
# 	‚Ä¢ Print all rows where City is "NY"

# Basic DataFrame Methods

# df.head(); # Gives the first 5 rows by default if the rows are less than five than all the rows
# df.info(); # Gives the information about the dataframe like data type , no of column etc.
# avg_age = df['Age'].mean() # gives the mean of a specific numeric colums
# print(avg_age)
# mode_age =  df['Age'].mode(); # gives mode
# print("Mode",mode_age)
# median_age = df['Age'].median() # gives the median of a specific numeric colums
# print(median_age)
# df.describe() # gives all the statistics calculation for all the numeric columns of the dataframe
#value_counts() # gives the count of category (like employees of each department)
#df['Department'].value_counts()


```

####	SQL Basics

	SQL Basics for Absolute Beginners
	
	What is SQL?
		‚Ä¢ SQL (Structured Query Language) is the language used to talk to databases.
		‚Ä¢ It lets you retrieve, add, update, and delete data stored in tables.
		‚Ä¢ Tables are like spreadsheets ‚Äî rows and columns of data.
	
	Common SQL commands you‚Äôll use:
		‚Ä¢ SELECT ‚Äî get data from a table
		‚Ä¢ FROM ‚Äî specify the table
		‚Ä¢ WHERE ‚Äî filter rows based on conditions
		‚Ä¢ GROUP BY ‚Äî group rows to do aggregations
		‚Ä¢ ORDER BY ‚Äî sort the results
	
	Example Table: Employees
	id	name	age	department	salary
	1	Alice	30	HR	50000
	2	Bob	25	IT	60000
	3	Charlie	35	IT	70000
	4	Diana	28	Marketing	52000
	
	1. Retrieve all columns and rows
	SELECT * FROM Employees;
		‚Ä¢ * means all columns
		‚Ä¢ This shows the whole table
	
	2. Retrieve specific columns
	SELECT name, salary FROM Employees;
		‚Ä¢ Only show name and salary columns
	
	3. Filter rows with WHERE
	SELECT * FROM Employees WHERE department = 'IT';
		‚Ä¢ Shows employees only in the IT department
	
	4. Filter rows with conditions
	SELECT * FROM Employees WHERE age > 28;
		‚Ä¢ Shows employees older than 28
	
	5. Sort results with ORDER BY
	SELECT * FROM Employees ORDER BY salary DESC;
		‚Ä¢ Sorts employees by salary from highest to lowest
	
	6. Group data with GROUP BY and aggregate with COUNT, AVG
	SELECT department, COUNT(*) AS num_employees, AVG(salary) AS avg_salary
FROM Employees
GROUP BY department;
		‚Ä¢ Shows each department, number of employees, and average salary
